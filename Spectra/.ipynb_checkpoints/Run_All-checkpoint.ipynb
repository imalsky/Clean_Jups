```python
phases = [0.0]
deltas = [0.0]

for phase in phases:
    for delta in deltas:
        plt.clf()
        df =  pd.read_csv('HayleyPlanet.txt',
                          delim_whitespace=True,
                          names=('lat', 'lon', 'level', 'alt', 'pres',
                                 'temp', 'u', 'v'), index_col = 'lat')

        # Get rid of any place where the longitude is equal to 360
        # Not needed by Eliza Code
        df = df[(df['lon'] != 360)].reset_index()

        orig_lats = df.lat
        orig_lons = df.lon

        def phase_rotation(df, phase):
            """ Rotate the planet a certain phase and rollover longitude"""
            # Rotate by the given phase
            df.lon = df.lon + phase

            # Rollover the longitude values if they
            # Are larger than 360.0

            df.lon = df.lon.mask(df.lon >= 360.0, df.lon - 360.0)
            return df


        def wind_rot(df):
            """ Calculate the new u wind speed"""
            u = df.u
            v = df.v
            lon = df.lon
            lat = df.lat

            # Convert to radians
            phi   = lat * (np.pi / 180.0)
            theta = lon * (np.pi / 180.0)

            # Get the prime cordinates
            phi_prime = np.arcsin(np.cos(delta) * np.sin(phi) -
                                  np.cos(theta) * np.sin(delta) * np.cos(phi))

            theta_prime = np.arctan2(np.sin(theta) * np.cos(phi),
                                     np.cos(delta) * (np.cos(theta) * np.cos(phi)) + np.sin(delta) * np.sin(phi))

            # Update the lat and lon with the new rotation
            lat_prime = phi_prime * (180.0 / np.pi)
            lon_prime = theta_prime * (180.0 / np.pi)

            # I am not including any w terms in the vector rotation
            v_x = u * np.sin(theta) + v * np.cos(theta) * np.sin(phi)
            v_y = -u * np.cos(theta) + v * np.sin(theta) * np.sin(phi)
            v_z = v * np.cos(phi)

            # All this vector math is in Deryl's google doc
            v_x_prime = -v_x * np.cos(delta) - v_z * np.sin(delta)
            v_y_prime = -v_y
            v_z_prime = v_x * np.sin(delta) + v_z * np.cos(delta)

            u_prime = (-v_x_prime) * np.sin(theta_prime) + v_y_prime * np.cos(theta_prime)
            v_prime = np.sin(phi_prime) * ((-v_x_prime * np.cos(theta_prime)) - v_y_prime * np.sin(theta_prime)) + v_z_prime * np.cos(phi_prime)

            # Update the dataframe
            df.lat = np.round(lat_prime, 2)
            df.lon = np.round(lon_prime, 2)
            df.u   = np.round(u_prime, 2)
            df.v   = np.round(v_prime, 2)
            return df

        df = phase_rotation(df, phase)
        df = df.apply(wind_rot, axis=1)
        df = phase_rotation(df, 180)

        levels = np.linspace(1, 60, 60)
        running_df = pd.DataFrame(columns=['lon', 'lat', 'temp', 'pres', 'alt', 'u', 'v'])

        for level in levels:
            print ('level')
            sub_df = df[(df['level'] == level)].reset_index(drop=True)
            sub_df = sub_df[(sub_df['temp'] > 100)]

            x = np.array(list(sub_df.lon))
            y = np.array(list(sub_df.lat))

            t2 = np.round(np.linspace(100, 100, 50), 3)
            t1 =  np.round(np.linspace(-10, 370, 50), 3)
            # It worked before with 100, 100
            xx, yy = np.meshgrid(t1, t2)

            columns = t1.astype(np.str)
            rows = t2.astype(np.str)

            rbf = Rbf(x, y, np.array(list(sub_df.temp)), epsilon=2, function='linear', smooth=1)
            z1 = rbf(xx, yy)
            z1[z1 < 100] = 0

            rbf = Rbf(x, y, np.array(list(sub_df.pres)), epsilon=2, function='linear', smooth=1)
            z2 = rbf(xx, yy)

            rbf = Rbf(x, y, np.array(list(sub_df.alt)), epsilon=2, function='linear', smooth=1)
            z3 = rbf(xx, yy)

            rbf = Rbf(x, y, np.array(list(sub_df.u)), epsilon=2, function='linear', smooth=1)
            z4 = rbf(xx, yy)

            rbf = Rbf(x, y, np.array(list(sub_df.v)), epsilon=2, function='linear', smooth=1)
            z5 = rbf(xx, yy)

            df_temp1 = pd.DataFrame(data=z1, index=rows, columns=columns)
            temp = df_temp1.stack().reset_index().rename(columns={"level_0": "lat", "level_1": "lon", 0: "temp"})

            df_temp2 = pd.DataFrame(data=z2, index=rows, columns=columns)
            pres = df_temp2.stack().reset_index().rename(columns={"level_0": "lat", "level_1": "lon", 0: "pres"})

            df_temp3 = pd.DataFrame(data=z3, index=rows, columns=columns)
            alt = df_temp3.stack().reset_index().rename(columns={"level_0": "lat", "level_1": "lon", 0: "alt"})

            df_temp4 = pd.DataFrame(data=z4, index=rows, columns=columns)
            u = df_temp4.stack().reset_index().rename(columns={"level_0": "lat", "level_1": "lon", 0: "u"})

            df_temp5 = pd.DataFrame(data=z5, index=rows, columns=columns)
            v = df_temp5.stack().reset_index().rename(columns={"level_0": "lat", "level_1": "lon", 0: "v"})

            big_df = pd.merge(temp, pres,  how='left')
            big_df = pd.merge(big_df, alt,  how='left')
            big_df = pd.merge(big_df, u,  how='left')
            big_df = pd.merge(big_df, v,  how='left')
            big_df['level'] = level
            print ('here')

            frames = [running_df, big_df]
            running_df = pd.concat(frames)

        """
        plt.pcolor(xx, yy, z1)
        plt.colorbar()
        plt.ylim(-100,100)
        plt.xlim(-10,370)

        plt.savefig('FIGS/phase_{}_delta_{}.pdf'.format(phase, delta), dpi=500)
        """
        
        running_df['lat'] = running_df['lat'].astype(float)
        running_df['lon'] = running_df['lon'].astype(float)
        running_df['level'] = running_df['level'].astype(float)

        running_df = running_df[['lat', 'lon', 'level', 'alt', 'pres', 'temp', 'u', 'v']]
        running_df = running_df.sort_values(by=['lat', 'lon'], axis=0, ascending=[True, True])

        running_df.to_csv('DATA/phase_{}_delta_{}.txt'.format(phase, delta), index=False, sep=' ')
```

```python
columns_to_add = ['c11', 'c12', 'c13', 'c14', 'c15', 'c16', 'c17', 'c18',
                  'c19', 'c110', 'c111', 'c112', 'c113']

input_paths = []              
output_paths = []       

for phase in phases:
    for delta in deltas:
        phase = str(phase)
        delta = str(delta)
        data_file = 'DATA/phase_' + phase + '_delta_' + delta + '.txt'
        df = pd.read_csv(data_file,  delim_whitespace=True)

        # Add the 0s
        for column in columns_to_add:
            df[column] = 0

        # Double the data
        double = df.copy()
        double.lon = double.lon + 360.0

        # Get rid of the lon = 360 values
        # I don't know why Eliza's code needs this
        double = double[(double['lon'] != 360)]
        doubled = pd.concat([df, double])

        doubled = doubled.sort_values(by=['lat', 'lon'])
        numpy_df = doubled.to_numpy()

        input_paths.append('Data/Final_{}_{}.dat'.format(phase, delta))
        np.savetxt('Data/Final_{}_{}.dat'.format(phase, delta), numpy_df,
        fmt='%.5E\t%.5E\t%i\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E\t%.5E')
    

inputs_file = 'input.h'

# The output paths should be similar to the input paxxxsths
# Minus the .dat file extension and saved to OUT/
for file_path in input_paths:
    output_paths.append('OUT/Doppler_No_Rotation_' + file_path[5:-4])


# Each Run needs to have a specific input.h file
# With the correct input and output paths
for i in range(len(input_paths)):
    output_temp = output_paths[i]
    input_temp  = input_paths[i]
    
    # Copy the template for inputs
    try:
        copyfile('template_inputs.h', inputs_file)
    except IOError as e:
        print("Unable to copy file. %s" % e)
        exit(1)
    except:
        print("Unexpected error:", sys.exc_info())
        exit(1)
    
    # Read in the file
    with open(inputs_file, 'r') as file :
        filedata = file.read()

    # Replace the input and output paths
    filedata = filedata.replace("<<output_file>>", "\"" + output_temp + "\"")
    filedata = filedata.replace("<<input_file>>", "\"" + input_temp + "\"")

    # Write the file out again
    with open(inputs_file, 'w') as file:
        file.write(filedata)
    
    # Run Eliza's code
    os.system('make clean')
    os.system('make rt_emission_aerosols.exe')
    os.system('./rt_emission_aerosols.exe')
    
    
    
```
